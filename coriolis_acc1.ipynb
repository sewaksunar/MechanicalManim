```python
from sympy.physics.vector import *
from sympy import symbols
from sympy import Function

O = ReferenceFrame('O')
P = ReferenceFrame('P')
t = symbols('t')
theta = Function('theta')(t)

P.orient(O, 'Axis', [theta, O.z])
```

$\displaystyle r_{S/O} \text{ expressed in } P = (u{\left(t \right)} + x_{P/O}{\left(t \right)} \cos{\left(\theta{\left(t \right)} \right)} + y_{P/O}{\left(t \right)} \sin{\left(\theta{\left(t \right)} \right)})\mathbf{\hat{p}_x} + (v{\left(t \right)} - x_{P/O}{\left(t \right)} \sin{\left(\theta{\left(t \right)} \right)} + y_{P/O}{\left(t \right)} \cos{\left(\theta{\left(t \right)} \right)})\mathbf{\hat{p}_y} + (w{\left(t \right)} + z_{P/O}{\left(t \right)})\mathbf{\hat{p}_z}$

```python
from sympy import latex
from sympy.physics.mechanics import dynamicsymbols
from IPython.display import Math

r_S_P, r_P_O = symbols(r'r_{S/P} r_{P/O}')
x, y, z = dynamicsymbols(r'x_{P/O} y_{P/O} z_{P/O}')
u= Function('u')(t)
v= Function('v')(t)
w= Function('w')(t)
r_P_O = x * O.x + y * O.y + 0 * O.z
r_S_P = u * P.x + v * P.y + 0 * P.z
r_S_O = r_S_P + r_P_O

display(Math(r"r_{S/O} \text{ expressed in } P = " + latex(r_S_O.express(P))))

v_S_O = r_S_O.diff(t, O)
display(Math(r"\frac{d}{dt}r_{S/O}\bigg|_{\text{represented in } P} = " + latex(v_S_O.express(P))))

# display(P.x.express(O))

A_S_O = v_S_O.diff(t, O)
display(Math(r"\frac{d^2}{dt^2}r_{S/O}\bigg|_{\text{represented in } P} = " + latex(A_S_O.express(O))))
```
